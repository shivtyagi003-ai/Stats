1-  What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?
- Type I error means you think something is true when it’s actually false.
Type II error means you miss something that’s actually true.
Type I is like a false alarm, while Type II is like overlooking a real problem.
These errors matter because they affect the decisions we make based on data.

2- What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?
- The p-value tells you how likely your sample results are if the null hypothesis were actually true.
A small p-value means your results are unlikely to be due to chance, so you tend to reject the null hypothesis.
A large p-value means the results are believable under the null, so you fail to reject it.
It does NOT prove the null is true or false—just measures evidence against it.

3- Explain the difference between a Z-test and a T-test, including when to use
each.
- A Z-test is used when the sample size is large or the population standard deviation is known.
A T-test is used when the sample size is small and the population standard deviation is unknown.
The T-test accounts for extra uncertainty in small samples.
Both compare sample results to what we'd expect under the null hypothesis.
In simple terms: Big sample = Z-test, Small sample = T-test.

4- What is a confidence interval, and how does the margin of error influence
its width and interpretation?
- A confidence interval  is a range of values that we believe, with a certain level of confidence (like 95%), contains the true population value.
Instead of giving one estimate (like the mean), it gives a range to show uncertainty.
How does the margin of error influence it?
- The margin of error (ME) is how much we allow our estimate to vary.

- A larger margin of error makes the confidence interval wider, meaning we are less precise but more confident.

- A smaller margin of error makes the interval narrower, meaning we are more precise but slightly less confident.

- Margin of error depends on sample size (bigger sample = smaller ME) and variability.

5- Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?
- ANOVA (Analysis of Variance) is used to compare the means of three or more groups to see if at least one group is significantly different from the others.

- If you were comparing only two groups, a t-test would be enough.

- But when there are 3, 4, or more groups, running multiple t-tests increases the chance of false positives (Type I error).

- ANOVA solves this by testing all groups at once with a single statistical test.
How ANOVA Extends Hypothesis Testing?
ANOVA uses the same logic as a t-test but for multiple groups:

- Null Hypothesis (H₀): All group means are equal.
(No difference between groups.)

- Alternative Hypothesis (H₁): At least one group mean is different.



```python
#Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
import math
import statistics as stats


data = [52, 49, 50, 48, 51, 53, 50, 49, 52, 47]


population_mean = 50


population_std = 3


sample_mean = stats.mean(data)
sample_size = len(data)


z_stat = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))


p_value = 2 * (1 - (0.5 * (1 + math.erf(abs(z_stat) / math.sqrt(2)))))

sample_mean, z_stat, p_value
```

```python
#Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
data = np.random.exponential(scale=2, size=100000)


sample_means = []
sample_size = 50
num_samples = 1000

for _ in range(num_samples):
    sample = np.random.exponential(scale=2, size=sample_size)
    sample_means.append(np.mean(sample))


plt.hist(data, bins=40)
plt.title("Original Distribution (Exponential)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()


plt.hist(sample_means, bins=40)
plt.title("Distribution of Sample Means (CLT Demonstration)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

sample_means[:10]
```

```python
#Write a Python function to calculate and visualize the confidence interval for a sample mean.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


data = np.array([12, 15, 14, 16, 18, 17, 19, 13, 14, 15])

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2, len(data) - 1)
    return mean, mean - margin, mean + margin

mean, lower, upper = confidence_interval(data)


print("Sample Data:", data)
print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")


plt.axvline(mean)
plt.axvspan(lower, upper, alpha=0.3)
plt.title("Confidence Interval Visualization")
plt.show()
```
